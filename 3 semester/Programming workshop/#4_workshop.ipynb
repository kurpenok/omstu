```python
from math import sin, cos

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform
```

```python
def f(x: float) -> float:
    return sin(2 * x) / cos(x)
```

```python
a = 2
b = 3
x0 = 2

array = np.arange(a, b, 0.05)
```

```python
array = np.array(list(map(f, array)))
```

```python
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Distribution of function values")

plt.plot(array, label=r"$\frac{sin(2x)}{cos(x)}$")
plt.legend()

plt.show()
```

```python
x = [i for i in range(100)]
y = [f(x[i]) for i in range(100)]

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Distribution of function values")
plt.grid(color = 'k', linestyle = ':')
plt.scatter(x, y, c="blue")
plt.show()
```

```python
sample_norm = norm.rvs(size=100)

plt.hist(sample_norm, density=True, bins=30)
plt.xlabel("Data")
plt.ylabel("Probability")
```

```python
def f(x1: float, x2: float) -> float:
    return (x1 - 3)**2 + (x2 - 8)**2


x = np.arange(10)
y = np.arange(10)

X, Y = np.meshgrid(x, y)

zs = np.array(f(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)
```
