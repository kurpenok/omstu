```python
import pandas as pd
import numpy as np
```

```python
data = [
    ["Strom, Mrs. Wilhelm (Elna Matilda Persson)", 29, "female"],
    ["Navratil, Mr. Michel (\"Louis M Hoffman\")", 36.5, "male"],
    ["Minahan, Miss. Daisy E", 33, "female"],
]

df = pd.DataFrame(data, columns=["Name", "Age", "Sex"])

df
```

```python
titanic_csv_df = pd.read_csv("data/titanic_csv.csv", sep=";")
titanic_csv_df.columns = map(str.lower, titanic_csv_df.columns)

titanic_csv_df
```

```python
url = "https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt"

titanic_url_df = pd.read_csv(url)

titanic_url_df
```

```python
titanic_df = pd.concat([titanic_csv_df, titanic_url_df])

titanic_df = titanic_df.drop_duplicates()

titanic_df
```

```python
titanic_df = titanic_df.set_index("passengerid")

titanic_df
```

```python
titanic_df = titanic_df.sort_index()

titanic_df
```

```python
titanic_df.info
```

```python
titanic_df.describe
```

```python
titanic_df.iloc[0], titanic_df.loc[2] = titanic_df.loc[2], titanic_df.iloc[0]

titanic_df
```

```python
titanic_df = titanic_df.replace({"sex": {"male": "m", "female": "f"}})

titanic_df
```

```python
titanic_df.groupby(by="ticket").filter(lambda tickets: len(tickets) >= 6)
```

```python
# Для выполнения 10 задания требуется наличие столбца с каютой в первом датафрейме
```

```python
titanic_df["birthyear"] = np.random.randint(1812, 1912, titanic_df.shape[0])

titanic_df
```

```python
companions = titanic_df.groupby(by="cabin")["name"].count().map(lambda x: x - 1)

titanic_df["companions"] = [None if (passenger["cabin"] != passenger["cabin"]) else companions[passenger["cabin"]] for _, passenger in titanic_df.iterrows()]

titanic_df
```

```python
titanic_df.iloc[0], titanic_df.loc[1] = titanic_df.loc[1], titanic_df.iloc[0]

titanic_df
```

```python
titanic_df.to_csv("data/new_titanic_df.csv", encoding="utf-8")
```
